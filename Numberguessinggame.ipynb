## Number Guessing Game (Console Version)


```python
import random # We'll need this to pick our secret number

def play_console_game():
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number...")

    # Initialize game parameters for the first round
    min_num = 1
    max_num = 100
    max_attempts = 7
    round_number = 1

    while True: # This loop keeps the game going for multiple rounds
        print(f"\n--- Round {round_number} ---")
        print(f"Guess a number between {min_num} and {max_num}. You have {max_attempts} attempts.")

        secret_number = random.randint(min_num, max_num) # The magic number!
        attempts_taken = 0
        guessed_correctly = False

        while attempts_taken < max_attempts: # The main guessing loop for a single round
            try:
                guess_input = input("Enter your guess: ")
                player_guess = int(guess_input) # Try to convert input to an integer

                # Check if the guess is within the current valid range (optional but good practice)
                if not (min_num <= player_guess <= max_num):
                    print(f"Please guess a number *between* {min_num} and {max_num}.")
                    continue # Don't lose an attempt for an out-of-range guess

                attempts_taken += 1 # Only count valid guesses as an attempt

                if player_guess < secret_number:
                    print("Too low!")
                elif player_guess > secret_number:
                    print("Too high!")
                else:
                    print(f"Congratulations! You guessed the number {secret_number} in {attempts_taken} attempts!")
                    guessed_correctly = True
                    break # Exit the inner loop, they won!

            except ValueError: # Catch if the user didn't enter a number
                print("Invalid input. Please enter a whole number.")
                # No attempt is lost for invalid input, loop continues

        if not guessed_correctly: # If they ran out of attempts
            print(f"Game Over! You ran out of attempts. The secret number was {secret_number}.")

        # Bonus: Increase difficulty and ask to play again
        play_again = input("\nDo you want to play again? (yes/no): ").lower()
        if play_again != 'yes':
            print("Thanks for playing! Goodbye!")
            break # Exit the outer loop, user doesn't want to play anymore
        else:
            # Make the next round harder!
            min_num = max(1, min_num - 20) # Widen the range a bit
            max_num = max_num + 20 # Widen the range more aggressively
            max_attempts = max(3, max_attempts - 1) # Maybe give one less attempt, but not fewer than 3
            round_number += 1
            print("Getting tougher now...")

# Let's run the console game!
play_console_game()
```

## Number Guessing Game (GUI Version with Tkinter)

```python
import tkinter as tk # This is the standard way to import Tkinter
from tkinter import messagebox # For nice pop-up messages
import random

class GuessingGameGUI:
    def __init__(self, master):
        self.master = master
        master.title("Number Guessing Game")
        master.geometry("400x300") # Set a decent initial window size

        self.min_num = 1
        self.max_num = 100
        self.max_attempts = 7
        self.attempts_taken = 0
        self.secret_number = 0 # Will be set at the start of each round
        self.round_number = 0

        self.create_widgets() # Build the GUI elements
        self.start_new_round() # Start the very first round

    def create_widgets(self):
        # Main instruction label
        self.instruction_label = tk.Label(self.master, text="", font=("Arial", 12))
        self.instruction_label.pack(pady=10)

        # Entry field for player's guess
        self.guess_entry = tk.Entry(self.master, width=20, font=("Arial", 14))
        self.guess_entry.pack(pady=5)
        self.guess_entry.bind("<Return>", self.check_guess_event) # Allows pressing Enter to submit

        # Guess button
        self.guess_button = tk.Button(self.master, text="Guess", command=self.check_guess, font=("Arial", 12))
        self.guess_button.pack(pady=5)

        # Feedback label (Too high/low/correct)
        self.feedback_label = tk.Label(self.master, text="", font=("Arial", 12, "italic"), fg="blue")
        self.feedback_label.pack(pady=5)

        # Attempts remaining label
        self.attempts_label = tk.Label(self.master, text="Attempts: ", font=("Arial", 10))
        self.attempts_label.pack(pady=5)

    def start_new_round(self):
        self.round_number += 1
        self.secret_number = random.randint(self.min_num, self.max_num)
        self.attempts_taken = 0

        # Reset GUI elements for a new round
        self.feedback_label.config(text="")
        self.guess_entry.delete(0, tk.END) # Clear any previous input
        self.guess_button.config(state=tk.NORMAL) # Enable the guess button
        self.guess_entry.config(state=tk.NORMAL) # Enable the entry field
        self.update_labels()

        messagebox.showinfo("New Round!",
                            f"Starting Round {self.round_number}!\nI've picked a new number between {self.min_num} and {self.max_num}.\nYou have {self.max_attempts} attempts.")

    def update_labels(self):
        # Update the instruction and attempts labels
        self.instruction_label.config(text=f"Round {self.round_number}: Guess a number between {self.min_num} and {self.max_num}!")
        self.attempts_label.config(text=f"Attempts remaining: {self.max_attempts - self.attempts_taken}")

    def check_guess_event(self, event):
        # This method is called when Enter is pressed in the entry field
        self.check_guess()

    def check_guess(self):
        try:
            guess_str = self.guess_entry.get()
            player_guess = int(guess_str)
            self.guess_entry.delete(0, tk.END) # Clear the entry after reading

            # Basic validation for range (optional, but good UX)
            if not (self.min_num <= player_guess <= self.max_num):
                self.feedback_label.config(text=f"Please enter a number between {self.min_num} and {self.max_num}.", fg="orange")
                return # Don't count this as an attempt

            self.attempts_taken += 1

            if player_guess < self.secret_number:
                self.feedback_label.config(text="Too low! Try again.", fg="red")
            elif player_guess > self.secret_number:
                self.feedback_label.config(text="Too high! Try again.", fg="red")
            else:
                self.feedback_label.config(text=f"Correct! You guessed {self.secret_number}!", fg="green")
                self.handle_win()
                return # Exit early after winning

            self.update_labels()

            if self.attempts_taken >= self.max_attempts:
                self.handle_loss()

        except ValueError:
            self.feedback_label.config(text="Invalid input. Please enter a whole number.", fg="orange")
            # No attempt lost for invalid input

    def handle_win(self):
        self.disable_game_inputs() # Stop further guesses
        response = messagebox.askyesno("You Won!",
                                        f"Congratulations! You guessed the number {self.secret_number} in {self.attempts_taken} attempts.\n\nDo you want to play another round? (Difficulty will increase!)")
        if response:
            self.increase_difficulty()
            self.start_new_round()
        else:
            self.master.quit() # Close the window if they don't want to play

    def handle_loss(self):
        self.disable_game_inputs() # Stop further guesses
        response = messagebox.askyesno("Game Over!",
                                        f"You ran out of attempts! The number was {self.secret_number}.\n\nDo you want to play another round? (Difficulty will increase!)")
        if response:
            self.increase_difficulty()
            self.start_new_round()
        else:
            self.master.quit() # Close the window

    def disable_game_inputs(self):
        self.guess_entry.config(state=tk.DISABLED)
        self.guess_button.config(state=tk.DISABLED)

    def increase_difficulty(self):
        # Make the next round harder, as per bonus requirement
        self.min_num = max(1, self.min_num - 20)
        self.max_num = self.max_num + 20
        self.max_attempts = max(3, self.max_attempts - 1) # Ensure at least 3 attempts
        # A little message for the user that it's getting harder
        messagebox.showinfo("Difficulty Up!", f"Next round: Numbers between {self.min_num} and {self.max_num}, with {self.max_attempts} attempts!")

# --- Main part to run the GUI ---
# This is what kicks off the Tkinter application
if __name__ == "__main__":
    root = tk.Tk() # Create the main window
    game = GuessingGameGUI(root) # Instantiate our game
    root.mainloop() # Start the Tkinter event loop, which keeps the window open and responsive
```
